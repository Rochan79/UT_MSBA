<h2>Problem 3</h2>

```python

rows=20
columns=20
A = np.zeros((20,20))
for i in range(rows):
    for j in range(columns):
        if i>j:
            A[i][j]=(j+1)/(i+1)
        else:
            A[i][j]=(i+1)/(j+1)

B=A.transpose()

def areSame(A,B):

    for i in range(rows):
        for j in range(columns):
            if (A[i][j] != B[i][j]):
                return 0
    return 1

areSame(A,B)
```

A is symmetric
