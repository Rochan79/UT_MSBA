# Question 1: Expected Loss Minimization [10 pts]
Consider a binary classification problem with the following loss matrix -
$$
   {\begin{array}{ccccc}
   & & \text{Predicted class} & \text{           } &\\
   & & C1 & C2 & Reject\\
   \text{True class} & C1 & 0 & r & c  \\
   & C2 & s & 0 & c \\
  \end{array} } 
$$

where the cost of rejection is a constant, and the costs $r$ and $s$ are positive real numbers. Let $f(x)=P(C1|x)$.


**(a) [2.5 points]** Show that the expected loss when $x$ is labelled as $C_1$ is a decreasing function of $f(x)$ while expected loss when $x$ is labelled as $C_2$ is a increasing function of $f(x)$.  

**(b) [2.5 points]** For $c=0$, show that the decision which minimizes the expected loss is to reject all instances of $x$

**(c) [2.5 points]** Let $r=5$ and $s=2$, what is the minimum value of $c$ such that no instance of $x$ gets rejected (irrespective of  $f(x)$)?

**(d) [2.5 points]** Let $r=7$, $s=4$, and $c=3$. Determine the  ranges of $f(x)$ for which the optimal decision is C1, reject and C2 respectively.

### Q1(a) Show that the expected loss when $x$ is labelled as $C_1$ is a decreasing function of $f(x)$ while expected loss when $x$ is labelled as $C_2$ is a increasing function of $f(x)$.  

$$
\text{Loss for accepting class }C_1\text{ is given by}\\
L(C_1 )= 0*f(x) +  s*(1-f(x))\\
= s(1-f(x))\\\\
\text{As we can see, the slope of the above line between Loss for choosing C1 and f(x) is negative, loss is a decreasing function of f(x)}\\\\
\text{Loss for accepting class }C_2\text{ is given by}\\
L(C_2 )= r*f(x) +  0*(1-f(x))\\
= rf(x)\\\\
\text{As we can see, the slope of the above line between Loss for choosing C2 and f(x) is positive, loss is a increasing function of f(x)}\\\\
$$


### Q1(b) For $c=0$, show that the decision which minimizes the expected loss is to reject all instances of $x$. 


$$
\text{We first see what loss we are getting in the first place}\\\\
\text{Expected Loss for choosing C1} = 0*f(x) + s*(1-f(x))\\
=s(1-f(x))\\\\
\text{As} s>0 \text{ }and \text{ } 1-f(x)>0\text{ }s(1-f(x))>0\\\\
\text{Similarly expected loss while choosing C2 =}rf(x)\\\\
\text{As both of them are always greater than 0 and the given loss to reject is 0, the loss to reject is better than the loss of choosing between any one of them}
$$


### Q1(c) Let $r=5$ and $s=2$, what is the minimum value of $c$ such that no instance of $x$ gets rejected (irrespective of  $f(x)$)?

$$
\text{Given r= 5 and s=2. Computing Loss associated with choosing C1 and C2}\\\\
L(C_1) = 0*f(x) + 2*(1-f(x))\\
=2-2f(x)\\
L(C_2) = 5*f(x) + 0*(1-f(x))\\
=5f(x)\\
\text{For no instance to be rejected the the loss of both classes can be equated such that,}\\\\
2-2f(x) = 5f(x)\\
f(x) = \frac{2}{7}\\
\text{for this f(x) the reject loss c would be}\\\\
c = 2(1-\frac{2}{7})\\
=\frac{10}{7}\\\\
\therefore \text{The minimum value of c such that no instance of x gets rejected is }c = \frac{10}{7}\\\\
$$


### Q1(d) Let $r=7$, $s=4$, and $c=3$. Determine the  ranges of $f(x)$ for which the optimal decision is C1, reject and C2 respectively.

First, we  find out at what value of f(x) does choosing either C1 or C2 have the same expected loss.

$$
4(1-f(x)) = 7f(x)\\
f(x) = 4/11\\
$$

This means that for $f(x) < 4/11$, it is better to choose C2 & vice versa for C1 (if we cannot reject)

At $f(x) = 4/11$, the expected loss is 

$$
\frac{28}{11} = 2.55\\
$$

Given that c=3, the expected loss of rejecting is more than the expected loss of just choosing. Hence, there is no 'reject' region.

We should choose C1 when $f(x) > 4/11$ & C2 when $f(x) < 4/11$

# Question 2: Logistic Regression [5 points]
Consider the problem of predicting the success of students in their undergraduate degrees given their high school GPA and SAT scores. We have the variables $X_1 =$ high school GPA, $X_2$ = SAT score and Y = completing the undergraduate degree (binary). We then fit a logistic regression which produces coefficients, $β_0 = −7$, $β_1 = 2$, $β_2 = 0.001$ Estimate the probability that George who
has a GPA of 3.4 and an SAT score of 1500 will complete his undergraduate degree successfully according to this model. By how much will the probability increase if the GPA increases to 3.7, with other factors remaining the same?

$$
ln(\frac{p}{1-p}) = \beta_0 + \beta_1X_1 + \beta_2X_2\\\\
\text{ 1. For the first part, inputting the values of given variables in the equation for logistic regression, we get }\\\\
ln(\frac{p}{1-p})  = -7 + 2*3.4 + 0.001*1500\\
\therefore ln(\frac{p}{1-p})  = 1.3\\
\therefore \frac{p}{1-p}  = e^{1.3}\\
\therefore  3.669 = \frac{p}{1-p}\\
\therefore p = 0.785 \approx 0.79\\\\
$$

Hence, chances that George completes his undergraduate degree is around 79%

$$
\text{2. For the next part of the question, we need to compute how much these odds change for changes in the GPA to 3.7}\\\\
\therefore ln(\frac{p}{1-p})  = -7 + 2*3.7 + 0.001*1500\\
\therefore ln(\frac{p}{1-p})  = 1.9\\
\therefore \frac{p}{1-p}  = e^{1.9}=6.685\\
\therefore  p = 0.869 \approx 0.87\\
\text{Increase in chances } = 0.87-0.79 = 0.08\\
$$
Hence, we can say that George's chances of completing the degree go up by almost 8% if the GPA increases to 3.7

```python
#Making a class for Confusion matrix as we have to make one again and again. 
class classification_metrics:
    def __init__(self, y_test, y_pred_test , y_valid , y_valid_pred, method):
        self.y_test = y_test
        self.y_pred_test = y_pred_test
        self.y_valid = y_valid
        self.y_valid_pred = y_valid_pred
        self.method = method
        
    def confusionMatrix(self, y_test,y_pred_test):
        confusion_matrix = metrics.confusion_matrix(y_test,y_pred_test)
        test_matrix = pd.DataFrame(confusion_matrix)
        return test_matrix
    
    def plot_confusionMatrices(self, y_test, y_pred_test , y_valid , y_valid_pred, method):
        test_matrix = self.confusionMatrix(y_test.astype(int), y_pred_test.astype(int))
        val_matrix = self.confusionMatrix(y_valid , y_valid_pred)
        fig, ax = plt.subplots(figsize=(15,8), nrows=1, ncols=2)
        plt.subplot(1,2,1)
        sns.set(font_scale=2)
        sns.heatmap(test_matrix, annot=True, fmt="g", cmap="magma")
        plt.title('Confusion Matrix of '+method+' : Test')
        plt.xlabel("Predicted label",fontsize = 13)
        plt.ylabel("True Label",fontsize = 13)

        plt.subplot(1,2,2)
        sns.set(font_scale=2)
        sns.heatmap(val_matrix, annot=True, fmt="g", cmap="magma")
        plt.title('Confusion Matrix of '+method+' : Validation')
        plt.xlabel("Predicted label",fontsize = 13)
        plt.ylabel("True Label", fontsize = 13)

        plt.show()
        print('Validation set F1 score:', round(metrics.f1_score(y_valid, y_valid_pred),2))
        print('Test set F1 score:', round(metrics.f1_score(y_test, y_pred_test),2))
         
```

# Question 5: Support Vector Machines [10 pts]

1. [3 pts] Explain in what sense the Support Vector Machine seeks an optimal answer in the space of linear classifiers.

Support Vector Machine is a linear model for classification. The SVM algorithm creates a hyperplane in N-dimensional data space which acts the decision boundary for the data classes. The first step for the algorithm is to find an approximate hyperplane separating the classes. Next, to optimize it, the algorithm finds closest points to the plane from both the classes. These points are called the support vectors. The distance between plane and the support vectors is called the margin. SVM maximizes the margin to get the optimal hyperplane. This hyperplane with maximum margin is the optimal hyperplane as it creates the best separation in data and giving optimal classification for linearly separable data.

2. [3 pts] Describe how a slack variable works.

There may be circumstances when the margins are 'soft margins', i.e. there might be some overlap between the two classes that hyperplane is unable to identify which allow for some misclassifications. However, by adding slack variables and minimizing the total error, we may define a separating hyperplane. Slack variables can be either be positive or zero and will loosen the rigidity of the separating hyperplane.

The slack variables are defined as the ratio between distance from a training point to a marginal hyperplane and half of the margin. The slack variables will reduce error and the margin size will be small. The data points should not cross over the margin. The amount to which the restriction on each observation can be broken depends on the slack variable. In order to achieve the best classification, a slack variable will be optimized for each observation and minimal use is preferable.

3. [4 pts] Explain how the SVM framework can be extended in order to provide non-linear decision boundaries.

SVM tries to find a hyperplane that can maximize the margin and improve accuracy for classification, and can be done when the data is linearly separable. However, in case of non-linear decision boundaries, SVM is used to transform the data into higher dimensions using the idea of Kernels. SVM creates a hyperplane in the high dimension after converting data to a high dimension where it can be linearly separable. Because there exists no hyperplane in the low-dimensional space, we converted the data into the high-dimensional space to determine our decision boundary, or hyperplane. There can be different types of kernel based on the kind of relationships to be formed in higher dimensions.
