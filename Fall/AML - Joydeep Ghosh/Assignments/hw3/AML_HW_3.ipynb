e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
z_critical = 1.96  # 95% = 1.96 standard deviations

#Get population standard deviation
pop_stdev = np.std(data['SalePrice'])



#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("Confidence Interval : ",confidence_interval)
print("Population Mean : ",population_mean)
````

````python
np.random.seed(12)

sample_size = 500

intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    # compute z critical value
    z_critical = 1.96
    # compute population standard deviation     
    pop_stdev = np.std(data['SalePrice'])
    
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    
plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

### Solution:

The "allowable deviation" from the population that a sample must have in order to be regarded to have come from the same population is known as the confidence interval. Our hypothesis (that the sample represents the population) is rejected if the sample mean deviates too much from the mean.

The horizontal line in the image above depicts the population mean, while the vertical line indicates the sample mean and the sample's 1.96*standard deviation value.



Since the above picture shows the 95% confidence interval of the population, we observe that, on average, if we pick a 100 samples from the population in an unbiased and independent way, only ‚âà 5% of the samples (3 samples to be precise, in this case) lie "outside" the allowable deviation.
