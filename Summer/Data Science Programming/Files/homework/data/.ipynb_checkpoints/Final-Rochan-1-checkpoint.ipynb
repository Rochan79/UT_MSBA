### [Q13 6 points] Cities preferring one type of booze

For each Location City, do the following:
* Calculate the totals for Liquor, Wine, and Beer Receipts from that City.
* From these totals, figure out the highest and second-highest totals
    * For example, if a city has totals of 100K for Liquor, 200K for Wine, and 10K for Beer, then the highest total is 200K and the second-highest is 100K.
* Print out the cities where the highest total is greater than 50,000 AND the ratio of highest to second-highest is greater than 10.

These are the cities that specifically prefer one kind of booze.

```python
daaru=df.copy()
saari_daaru=pd.DataFrame(daaru.groupby('Location City')[['Liquor Receipts','Wine Receipts','Beer Receipts']].sum())
daaru_t = saari_daaru.T.copy()
saari_daaru['Second Highest'] = daaru_t.apply(lambda x: x.nlargest(2)[1])
saari_daaru['Highest'] = saari_daaru.max(axis=1)
saari_daaru['Ratio'] = saari_daaru['Highest']/saari_daaru['Second Highest']
saari_daaru[(saari_daaru['Highest']>50000) & (saari_daaru['Ratio']>10) & (saari_daaru['Second Highest']!=0)]
```

### [Q17 6 points] Pattern of places being within city limits

* For each Location City, figure out the number of rows associated with that City ("numRows"), and the fraction of these rows where the 'Inside/Outside City Limits' columns equals 'Y' ("fracY").
* Split the Location Cities into 5 bins based on their "numRows". Each bin should have roughly the same number of Location Cities in it.
* For each bin, print out the average "fracY" for all Location Cities in that bin.

```python
df['one']=1
n = len(df)
count_total = df.groupby(['Location City'])['Inside/Outside City Limits'].count()
Y_count = df[df['Inside/Outside City Limits']=='Y'].groupby(['Location City'])['one'].count()
pattern = pd.concat([count_total,Y_count],axis=1).rename(columns={'Inside/Outside City Limits':'numRows','one':'Yescount'})
pattern['fracY'] = pattern['Yescount']/pattern['numRows']

pattern['bin']=pd.qcut(pattern['numRows'],5,labels=['1st Bin','2nd Bin','3rd Bin','4th Bin','5th Bin'])
pattern.groupby(['bin'])['fracY'].mean()
```

* For each row, figure out which one of Liquor/Wine/Booze Receipts "dominates" (i.e., its value is  greater than the other 2 receipts)

    * If Liquor=Beer=100 and Wine=10, then no one dominates
    * If Liquor=Wine=Beer=0, then no one dominates.
* Note that each Taxpayer Name can show up multiple times (i.e., multiple rows), each time with a different type of booze dominating the others.
* Figure out the Taxpayer Names for which there are at least 2 rows in which Liquor dominates, at least 2 rows in which Wine dominates, and at least 2 rows in which Beer dominates
* Report the fraction of unique Taxpayer Names for which this happens.

```python
df_bev=df.copy()

def dominator(x):
    
    max_receipt = max(x['Liquor Receipts'], x['Wine Receipts'], x['Beer Receipts'])
    if (x['Liquor Receipts'] == x['Wine Receipts']) | (x['Wine Receipts'] == x['Beer Receipts']) | (x['Beer Receipts'] == x['Liquor Receipts']):
        return NaN
    else:
        if x['Liquor Receipts'] == max_receipt:
            return 'Liquor'
        elif x['Wine Receipts'] == max_receipt:
            return 'Wine'
        else:
            return 'Beer'
        
df_bev['Dominating daaru'] = df_bev.apply(dominator, axis = 1)

taxpayers = df_bev.groupby(['Taxpayer Name','Dominating daaru']).size()
rows_2 = taxpayers[taxpayer_cnts >= 2].groupby('Taxpayer Name').size().sort_values(ascending = False)
sum(rows_2 > 1)/len(df_bev['Taxpayer Name'].drop_duplicates())

```

### [Q19 8 points] Predictions for Cervical Cancer

Load in **cervical.csv**. Each row lists the details of one patient. 

* A patient is in the positive class ("has cancer") if any of the following four columns is greater than 0: Hinselmann/Schiller/Citology/Biopsy.

Set up the following classification problem:
* We want to predict if a patient is in the positive class using all features *except* the four columns named above.
* Split the data into 75% training and 25% testing set.
* Print out the test accuracy of the baseline classifier.
* Print out the test accuracy of a Gradient Boosting Regression classifier with 100 trees, each of depth 1.

```python
from patsy import dmatrices
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

df_modeling = pd.read_csv("cervical.csv")


df_modeling.columns = df_modeling.columns.str.replace(' ','_')
df_modeling.columns = df_modeling.columns.str.replace(':','_')
df_modeling.columns = df_modeling.columns.str.replace('-','_')
df_modeling.columns = df_modeling.columns.str.replace('__','_')
df_modeling.columns = df_modeling.columns.str.replace('/','')
df_modeling.columns = df_modeling.columns.str.replace('(','')
df_modeling.columns = df_modeling.columns.str.replace(')','')


df_modeling['prediction_class'] = 0
df_modeling.loc[(df_modeling['Hinselmann']>0) | (df_modeling['Biopsy'] > 0)  | (df_modeling['Citology'] > 0) | (df_modeling['Schiller'] > 0) ,'prediction_class'] = 1

formula = 'prediction_class ~ 0 + Age + Number_of_sexual_partners + First_sexual_intercourse + Num_of_pregnancies + '\
'C(Smokes) + Smokes_years + Smokes_packsyear + C(Hormonal_Contraceptives) + Hormonal_Contraceptives_years + '\
'C(IUD) + IUD_years + C(STDs) + STDs_number + C(STDs_condylomatosis) + C(STDs_cervical_condylomatosis) + '\
'C(STDs_vaginal_condylomatosis) + C(STDs_vulvo_perineal_condylomatosis) + C(STDs_syphilis) + C(STDs_pelvic_inflammatory_disease) + '\
'C(STDs_genital_herpes) + C(STDs_molluscum_contagiosum) + C(STDs_AIDS) + C(STDs_HIV) + C(STDs_Hepatitis_B) + C(STDs_HPV) + '\
' STDs_Number_of_diagnosis + C(Dx_Cancer) + C(Dx_CIN) + '\
'C(Dx) + C(Dx_HPV)'

Y, X = dmatrices(formula, df_modeling, return_type='dataframe')
y = Y['prediction_class'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

negative_examples_in_test = len(y_test[y_test==0])
total_examples_in_test = len(y_test)

print('Baseline accuracy:', negative_examples_in_test * 1.0 / total_examples_in_test)

gbr_params = {'n_estimators': 100,
          'max_depth': 1,
          'learning_rate': 0.2}

gbr = GradientBoostingClassifier(**gbr_params)

gbr.fit(X_train, y_train)

y_pred=gbr.predict(X_test)


print("Boosting accuracy:")
print(accuracy_score(y_test, y_pred))
```
