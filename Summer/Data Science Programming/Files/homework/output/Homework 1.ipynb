Welcome Message and menu functions defined below.

```python
def welcome_message(): 
    '''Function that prints a welcome message for user'''
    
    print("Wecome to Rock, Paper, Scissors! See if you can beat the computer!")
    print("The game is simple. The computer will choose first, keep it a secret," +
          " and allow you to select a choice from the following menu:")

    
def print_menu():
    '''Function that prints the menu options for the user'''
    
    print("\nHere are your menu options:" +
          "\n1. Rock\n2. Paper\n3. Scissors\n4. Quit\n")
```

User Input Validation function defined below.

```python
def get_valid_input():
    '''Function that validates the user's numerical input'''

    print_menu()   #calls the print_menu() function below
    choice = -1    #sets choice to arbitrary bad value to keep loop running until output valid
    while choice == -1: 
        choice = input("Choose a valid number from the menu: ")
        try: 
            choice = int(choice)
        except: 
            print("Error. Please Enter a valid number from the menu. Try again.")
            choice = -1
        else: 
            if choice != 1 and choice != 2 and choice != 3 and choice != 4:    #these are the valid choices
                print("Error. Please Enter a valid number from the menu. Try again.")
                choice = -1
    return choice    
```

Function to store human response information defined below.

```python
def store_human_responses(human_choice):
    '''Function that takes a numeric input marking human choice and stores it in a list that is returned 
    and passed to get_computer_response() function. It will serve as an internal tracker of human responses'''

    ROCK = 1       #placeholders for clarity
    PAPER = 2
    SCISSORS = 3
    
    
    if human_choice == ROCK:
        human_data[0] = human_data[0] + 1     #increment human_data at first index by one if user plays Rock
    elif human_choice == PAPER:
        human_data[1] = human_data[1] + 1     #increment human_data at first index by one if user plays Paper
    elif human_choice == SCISSORS:
        human_data[2] = human_data[2] + 1     #increment human_data at first index by one if user plays Rock

    return human_data
```

Function that determines the computer's response after the first randomly generated response.

```python
def get_computer_response(human_data):
    '''Function that takes the human_data list as input and uses the maximum frequency of plays to beat the human, 
    assuming that the human will continue to play its most common strategy. It needs to be called for rounds following 
    the first round ONLY. It returns the computer's response as an int.'''

    ROCK = 1        #placeholders for clarity
    PAPER = 2
    SCISSORS = 3
    
    computer_response = -1   #initialize variable to a known bad value. 

    num_rock = human_data[0]      #number of Rocks is first index position of human_data
    num_paper = human_data[1]
    num_scissors = human_data[2]
    
    if num_rock > num_paper:
        if num_paper > num_scissors:
            computer_response = PAPER                     #human goes rock, comp goes paper 
        elif num_rock > num_scissors:
            computer_response = PAPER                     #human goes rock, comp goes paper
        else:
            computer_response = ROCK                      #human goes scissors, comp goes rock
    else:
        if num_rock > num_scissors:
            computer_response = SCISSORS                  #human goes paper, computer goes scissors  
        elif num_paper > num_scissors:
            computer_response = SCISSORS                  #human goes paper, computer goes scissors 
        else:
            computer_response = ROCK                      #human goes scissors, computer goes rock 

    if num_rock == num_paper and num_rock > num_scissors:
        computer_response = random.randint(2,3)                #R and P tied at top, computer will go P or S (50/50)

    if num_rock == num_scissors and num_rock > num_paper:
        computer_response = random.randint(1,2)                #R and S tied at top, computer will go P or R (50/50)

    if num_paper == num_scissors and num_paper > num_rock:
        computer_response = random.randrange(1, 3, 2)          #P and S tied at top, computer will go S or R (50/50)
        
    if num_paper == num_rock == num_scissors:                  #unique case of triple equality 
        computer_response = random.randint(1,3)

    
    return computer_response
```

Function that plays the entire game and determines the winner. This is like the main() function.

```python
def determine_winner():
    '''Function that contains the mainline logic of the program and runs the game. It calls all other predefined
    functions above.'''
    
    #prints the welcome message for the user
    welcome_message()
    
    #gets a validated input from the user 
    human_choice = get_valid_input()  

    #first computer choice is random
    computer_choice = random.randint(1,3)
    
    #initialize variables to 0
    total_games = 0
    total_human_wins = 0
    
    #dictionary that stores the game plays according to the menu
    game_dictionary = {1: "Rock", 2: "Paper", 3:"Scissors"}

    
    #loops while human choice is not 4 (4 means quit)
    while human_choice != 4:
            
        #3 beats 2, 2 beats 2, 1 beats 3
        if human_choice == 2 and computer_choice == 1:
            total_human_wins += 1                 #increment human wins to track
            message = "Congrats! You won."        #winner message
        if human_choice == 3 and computer_choice == 2:
            total_human_wins += 1
            message = "Congrats! You won."
        if human_choice == 1 and computer_choice == 3:
            total_human_wins += 1
            message = "Congrats! You won."
        if human_choice == 1 and computer_choice == 2:
            message = "The computer won. Better luck next time!"
        if human_choice == 2 and computer_choice == 3:
            message = "The computer won. Better luck next time!"
        if human_choice == 3 and computer_choice == 1:
            message = "The computer won. Better luck next time!"


        if human_choice == computer_choice:      #if there is a tie, message is Tie Game
            message = "Tie Game!"

        total_games += 1                         #increment each time a game is played to track total games

        
        #message that prints a summary of your move and the computer's move along with Win/Loss/Tie message
        print("\nYour move:", game_dictionary[human_choice])
        print("Computer's move:", game_dictionary[computer_choice])
        print(message)
        
        #call the store_human_responses() function to store human data in running list
        human_data_list = store_human_responses(human_choice)
        #print(human_data_list)

        #determine the computer's choice sending human_data_list as parameter
        computer_choice = get_computer_response(human_data_list)

        #get human's choice to enter the loop again or exit playing
        human_choice = get_valid_input()
        

    #when loop is exited, game summary message printer for the user 
    print("\n--------------------GAME SUMMARY--------------------")
    if human_choice == 4:
        if total_games == 1:
            print("\nYou played ",total_games, " total game.", sep="")
        else: 
            print("\nYou played ",total_games, " total games.", sep="")
        if total_human_wins == 1:
            print("You won ", total_human_wins, " game.", sep="")
        else: 
            print("You won ", total_human_wins, " games.", sep="")
        print("Thanks for playing!")

        
#calling the determine_winner() function to run program
determine_winner() 
```

## Question 3: Near-duplicate detection

```python
import math
import regex as re
```

<b>Part A</b>

```python
def moving_window(tweet) :
    '''Take in a tweet and split it by spaces, creates three word phrases incrementing by one each time. 
    Returns a dictionary in the format key: phrases, values: 1'''
    tweet_dict = {}

    tweet_list = tweet.split(" ")

    for index in range(0, len(tweet_list)-2) :
        tweet_chunk_list = tweet_list[index:index+3]
        tweet_chunk_string = " ".join(tweet_chunk_list).rstrip()
        tweet_dict[tweet_chunk_string] = 1
    
    return tweet_dict

# Sample Output
moving_window("This is an awesome tweet, dude")
```

<b>Part B</b>

```python
def cosine(tweet_dict1, tweet_dict2) :
    '''Parameters are two tweet dictionaries keys: phrases, values: 1.  Function finds the cosine similarity
    score for the two tweets by searching for common phrases (matches/sqrt(n1*n2))'''
    num_phrase1 = len(tweet_dict1)
    num_phrase2 = len(tweet_dict2)

    matches = 0
    for key, value in tweet_dict1.items() :
        if key in tweet_dict2 :
            matches += 1
    
    n_root = math.sqrt(num_phrase1 * num_phrase2)

    return matches / n_root

# Sample Output
tweet_dict_one = moving_window("This is an awesome tweet, dude")
tweet_dict_two = moving_window("This is an awesome tweet, mark")
cosine(tweet_dict_one, tweet_dict_two)
```

<b>Part C</b>

```python
def match_tweets() :
    '''
        Opens and reads santa file of tweets and appends each tweet to a list.
        For each tweet in the list compare it to the unique combinations of tweets 5C2
        Calculate the score using the cosine function, after converting it to phrases using the moving window function
        If the cosine score > 0.5, output duplicate found and display the corresponding strings
    '''
    #open the Santa.txt file 
    santa_file = open("Santa.txt", "r")

    full_santa_list = []

    for line in santa_file: 
        full_santa_list.append(line.rstrip())

    santa_file.close()
    
    count = 0
    print()
    print("------------------------RESULTS---------------------------")
    for index in range(0,len(full_santa_list)) :
        for index2 in range(index + 1, len(full_santa_list)) :
            score = cosine(moving_window(full_santa_list[index]), moving_window(full_santa_list[index2]))
            if score >0.5 :
                count += 1
                print()
                print(str(count) + ". DUPLICATE FOUND:", "Tweet", index2 + 1,  "is similar to Tweet", index + 1)
                print()
                print("Cosine Similarity Score:", format(score, ",.4f"))
                print()
                print("Tweet ("+ str(index2+1) + "):", full_santa_list[index2]) # index2
                #print(full_santa_list[index2])
                print()
                print("Near Duplicate Tweet ("+ str(index+1) + "):", full_santa_list[index]) # index
                #print(full_santa_list[index])
                print()
                print()

    print("------------------------SUMMARY---------------------------")
    print()
    print("There were", count, "pairwise matches found out of", len(full_santa_list), "Tweets.")
    print()
    
match_tweets()
```

## Question 4: The Google of Quotes

```python
import regex as re
import math
```

<b>Part A</b>

```python
#encoding thing is to help w error when file was read 
quotes_file = open("quotes.txt", "r", encoding = "windows-1252")

master_list = []

quote = quotes_file.readline()

# read in file and create master list
while quote != "": 
    quote = quote.rstrip("\n")
    author = quotes_file.readline().rstrip("\n")

    combo_string = quote + " - " + author

    master_list.append(combo_string)

    #read in the next line; continue until "" read
    quote = quotes_file.readline()

quotes_file.close()

# Sample Output
master_list # list of the full quotes of the form "quote - speaker"
```

<b>Part B</b>

```python
def breakup_master_list(full_quote) : 
    '''Reads in a quote and splits it into a list of the words, returns the list of words'''
    single_quote_list = re.split("[^a-zA-Z0-9_]+", full_quote.lower())
    return(single_quote_list)

# Sample Output
quote = 'The heart has its reasons, of which the mind knows nothing. - Blaise Pascal'
breakup_master_list(quote)
```

<b>Part C</b>

```python
def get_word_values_dict(full_quote) : 
    '''
        Call the breakup_master function to get a list of the words in a single quote
        Set creates a unique list with only a single occurence of each word
        Loop through each word, find count using list returned from breakup_master_list function
            and comparing the words to the set of unique words, add total count to value_dict as the value,
            with word as the key
    '''
    value_dict = {}

    #call the breakup_master function to get indiv lists of component words 
    quote_words_list = breakup_master_list(full_quote)

    # set creates a unique list with only a single occurence of each word
    # loop through each word, find count using original list and set of unique words, add to value_dict
    for word in set(quote_words_list) :
        count = quote_words_list.count(word)
        value_dict[word] = count

    return value_dict

# Postings List Dictionary (part C)
post_list_dict = {}

for full_quote in master_list: 
    post_list_dict[full_quote] = get_word_values_dict(full_quote)

post_list_dict
```

<b>Part D</b>

```python
# Reverse Postings-list Dictionary (part D)
# 1 --> Create unique list of all words, by sending master_list quotes to break up function
all_words_list = []
for full_quote in master_list :
    quote_words_list = breakup_master_list(full_quote)
    for word in quote_words_list :
        all_words_list.append(word)

unique_words_list = list(set(all_words_list))
unique_words_list.remove('')

# 2 --> Iterate through the unqiue set of words and find where it's in the PLD dictionary, 
#       by searching through interior dictionaries keys (the words)
reverse_post_list_dict =  {}
for search in unique_words_list :
    single_word_dict = {}
    for quote, word_count_dict in post_list_dict.items() :
        for word, count in word_count_dict.items() :
            if search == word :
                single_word_dict[quote] = count
                reverse_post_list_dict[search] = single_word_dict


# loop through the master list and call breakup_master_list
# this generates a separate list for each quote + auth combo w indiv strings 
for full_quote in master_list :
    #call the breakup_master_list function
    split_quotes_list = breakup_master_list(full_quote)
    #print(split_quotes_list)


# Sample Output (reverse_post_list_dict data structure)   
reverse_post_list_dict
```

<b>Part E</b>

```python
# Get TF-IDF and return value (part E)
def get_tf_idf(word, quote, count) :
    '''
        TF - Term Frequency (occurences of the word / max occurences of most frequent word)
        IDF - Inverse Document Frequency ln(total number of quotes / number of quotes that include specific word)
        TF-IDF = TF * IDF 
        TF-IDF is returned by the function after using word, quote and count to calculate TF and IDF
    '''
    max_repeated_word = max(post_list_dict[quote].values())
    tf = count / max_repeated_word

    num_quotes_with_word = len(reverse_post_list_dict[word])

    idf = math.log(len(master_list)/num_quotes_with_word)
    return tf * idf

# Sample Output 
get_tf_idf("entertainer", "An actor is at most a poet and at least an entertainer. - Marlon Brando", 1)
```

<b>Part F</b>

```python
# Quote search using a single word (part F)
def quote_word_search(word) :
    '''
        Define a new dictionary for output quote_tf_idf_dict (keys: quotes, values: TF-IDF for quote)
        Loop through the dictionary from the reverse post_list_dict using key word
        For quote in loop, set quote_tf_idf_dict key equal to quote and get corresponding TF-IDF score for the value
        Return quote_tf_idf_dict
    '''
    # access R PLD key to get word we need
    quote_tf_idf_dict = {}
    if word in reverse_post_list_dict.keys() :
        quote_count_dict = reverse_post_list_dict[word]
        for quote, count in quote_count_dict.items() :
            # call TF-IDF to get scores
            quote_tf_idf_dict[quote] = get_tf_idf(word, quote, count)
        # return dict (key --> Quotes, values --> TF-IDF for quote)
    else :
        print(word,"is not found in this set of quotes!")
    
    return quote_tf_idf_dict


# Sample Output for "entertainer" and "dogs"
for word in ['entertainer', 'dogs'] :
    print(word)
    print(quote_word_search(word))
    print()
```

<b>Part G</b>

```python
# Quote search using multiple words (part G)
def quote_multi_word_search(words) :
    '''
        Takes a list of words as the parameter
        Define a new dictionary for output multi_quote_tf_idf_dict (keys: quotes, values: TF-IDF for quote(s))
        Loop through the dictionary from the reverse post_list_dict using key word
        For quote in loop, 
            Case 1: set quote_tf_idf_dict key equal to quote if quote isn't already in the 
            multi_quote_tf_idf_dict and get corresponding TF-IDF score for the value
            Case 2: if the quote is already in multi_quote_tf_idf_dict, index the dictionary and add the 
            corresponding TF-IDF score to the key, value pair 
        Return multi_quote_tf_idf_dict
    '''
    multi_quote_tf_idf_dict = {}
    for word in words :
        if word in reverse_post_list_dict.keys() :
            quote_count_dict = reverse_post_list_dict[word]
            for quote, count in quote_count_dict.items() :
                if quote not in multi_quote_tf_idf_dict.keys() :
                    multi_quote_tf_idf_dict[quote] = get_tf_idf(word, quote, count)
                else :
                    multi_quote_tf_idf_dict[quote] += get_tf_idf(word, quote, count)
        else :
            print(word,"is not found in this set of quotes!")

    return multi_quote_tf_idf_dict


# Sample Output
quote_multi_word_search(['heart', 'mind', 'disease'])
```
